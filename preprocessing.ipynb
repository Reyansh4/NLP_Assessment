```python
import pandas as pd
import os
import re
```

```python
data = pd.read_csv("data\data.csv")
data
```

```python
data['date'] = data['date'].str.replace(r'Published:\s*', '')
data
```

```python
data['editor'] = data['editor'].str.replace(r'Edited By:\s*','')
data
```

```python
data['heading'] = data['heading'].str.replace(r':','')
```

```python
data['heading'] = data['heading'].apply(lambda x: re.sub(r'\b[a-zA-Z]+\b','',x))
data
```

```python
data['news_text'] = data['news_text'].apply(lambda x: re.sub(r'\b[a-zA-Z]+\b', '',x))
data
```

```python
data['heading'] = data['heading'].apply(lambda x: re.sub(r'\b[0-9]+\b', '', x))
data['news_text'] = data['news_text'].apply(lambda x: re.sub(r'\b[0-9]+\b', '',x))
```

```python
data
```

```python
pattern = re.compile(r'^\s*(यह भी पढ़ें|ये भी पढ़ें).*$', flags=re.IGNORECASE)

data['news_text'] = data['news_text'].apply(lambda x: re.sub(pattern, '', x))
data
```

```python
pattern_1 = re.compile(r'[.,(){}\[\]<>?!@#$%^&*;:/\\|`~"\'=_+-]+')

data['news_text'] = data['news_text'].apply(lambda x: re.sub(pattern_1, '', x))

data['heading'] = data['heading'].apply(lambda x: re.sub(pattern_1, '', x))

data
```

```python
data['date'] = data['date'].str.strip('"')
data['date']
```

```python
data.to_csv("data\mod_data.csv", index=False)
```

```python
data['corpus'] = data['heading'] + data['news_text']
data['corpus']
```

```python
data[['corpus']].to_csv('covid_19_corpus.txt', index=False, header=False)
```
